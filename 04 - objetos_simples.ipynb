# Exercício 01:

No paradigma de Programação Orientada a Objetos, o que é um Objeto?
<br>
<br>
R.: Um objeto é uma entidade que possui atributos e métodos. Essa entidade é uma instância de uma classe. Por exemplo, "string" é um objeto do tipo string que tem características próprias (métodos e atributos) e os caracteres "string" são a instância (nome) desse objeto específico pertencente à classe string. 

# Exercício 02:

Em Python, o método com nome count() faz parte do objeto da classe string. Este objeto possuí outros atributos e métodos. Qual o código que pode ser usado para verificar se count() é um método também presente no objeto da classe integer?

```python
#Resposta:
(100).count(1) #Essa linha mostra o erro 'count' object has no attribute 'count', o que demonstra que o tipo 'int' não tem o método 'count'. 
dir(int) #Essa linha mostra os métodos disponíveis para o tipo 'int', nos quais count não está presente.
```
