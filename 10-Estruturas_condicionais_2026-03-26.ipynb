p and q and r and s and t?
<br>

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2
resultado = p and q and r and s and t
if resultado == False :
    print ("O resultado da operação é falso: ", p, q, r, s, t)
else :
  print ("O resultado da operação é verdadeiro", p, q, r, s, t)
```
