## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

Não usei uma estrutura de repetição, mas sim uma estrutura de seleção para separar os genes em pares e ímpares. Poderia ter feito assim, com repetição:

```python
import pandas as pd

df = pd.read_csv("df_genes.csv")
df = df.T

pares = 0
impares = 0

for gene in df.index:
    tamanho = len(gene)
    
    if tamanho % 2 == 0:
        pares += 1
    else:
        impares += 1

print("Pares:", pares)
print("Ímpares:", impares)
```
