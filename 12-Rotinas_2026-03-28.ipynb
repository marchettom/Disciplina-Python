## Responda a pergunta do exercício 2 da aula 10, usando funções:<br>
Instancie um objeto da classe NumPy array, contendo 5 valores quantitativos randômicos normalmente distribuídos, com média 10 e desvio padrão 3.<br><br>
Quantos e quais são os valores superiores a 10?

## 02 – Descreva o racional da sua resposta:

R.: O racional é basicamente o mesmo da atividade anterior. Contudo, a seleção dos valores acontece dentro da função (que pode ser chamada diversas vezes dentro de um mesmo código). Nesse caso ela retorna três valores diferentes que correspondem ao array de valores aleatórios, a quantidade de valores maiores que 10 e os valores em si.

## 04 – Escreva o programa em Python:

```python
import numpy as np

def maiores_que_10():

    arr = np.random.normal(loc=10, scale=3, size=5)
    
    
    valores_maiores = arr[arr > 10]
    
    
    quantidade = np.sum(arr > 10)
    
   
    return arr, quantidade, valores_maiores


array, qtd, maiores = maiores_que_10()

print(array)
print("Quantidade > 10:", qtd)
print( maiores)
```

## 05 – Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

Sim, bastaria mudar os parâmetros:

```python
import numpy as np

def maiores_que_20():

    arr = np.random.normal(loc=10, scale=3, size=5)
    
    
    valores_maiores = arr[arr > 20]
    quantidade = np.sum(arr > 20)
    
   
    return arr, quantidade, valores_maiores



array1, qtd1, maiores1 = maiores_que_20()

print(array1)
print("Quantidade > 20:", qtd1)
print( maiores1)
```

## 06 – Descreva o racional da sua resposta:

R.: Criei uma função em que a variável soma recebe a soma de um conjunto de valores (input). Em seguida é calculada a quantidade de valores e calculada a média dividindo a soma pela quantidade de valores. Fora da função, defini a variável "valores", que recebe o input. Esse input é convertido em uma lista de floats na linha 10. Por fim, o código chama a função e printa o valor retornado por ela.

## 08 – Escreva o programa em Python:

```python

def calcular_media(valores):
    soma = sum(valores)
    quantidade = len(valores)
    media = soma / quantidade
    return media

valores = input("digite os valores separados por espaço")

valores = [float(x) for x in valores.split()]

print (calcular_media(valores))
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir_objetos(obj1, obj2):
    print("Objeto 1:")
    print("Classe:", type(obj1))
    print("Valor:", obj1)
    
    print("Objeto 2:")
    print("Classe:", type(obj2))
    print("Valor:", obj2)

    if type(obj1) == str:
        return "Impossível dividir um objeto str."   #Tinha feito essa parte pensando em instanciar uma variável direto no código. Mas no input mesmo coloquei a conversão para float.
    
    if type(obj2) == str:
        return "Impossível dividir um objeto str."
    
    if type(obj1) == bool:
        return "Impossível dividir um objeto bool."
    
    if type(obj2) == bool:
        return "Impossível dividir um objeto bool."
    
    if obj2 == 0:
        return "Erro: divisão por zero não é permitida."
    
    resultado = obj1 / obj2
    return resultado

#a = 2
#b = 4

#a = True
#b = "string"

a = float(input("Digite o primeiro valor"))
b = float(input("Digite o segundo valor")) 

print("Resultado da divisão:", dividir_objetos(a, b))
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def dividir_compostos(obj1, obj2):
    print("Objeto 1:")
    print("Classe:", type(obj1))
    print("Valor:", obj1)
    
    print("Objeto 2:")
    print("Classe:", type(obj2))
    print("Valor:", obj2)
    

    if not isinstance(obj1, list) or not isinstance(obj2, list):   #verificando tipo
        return "Erro: os objetos devem ser listas."
    

    if len(obj1) != len(obj2):
        return "Erro: listas devem ter o mesmo tamanho."  #verificando tamanho
    
    resultado = []
    
    for a, b in zip(obj1, obj2):
        if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
            return "Erro: todos os elementos devem ser numéricos."
        
        if b == 0:
            return "Erro: divisão por zero em um dos elementos."
        
        resultado.append(a / b)
    
    return resultado


lista1 = [10, 20, 30]
lista2 = [2, 4, 5]

print("Resultado:", dividir_compostos(lista1, lista2))
```
